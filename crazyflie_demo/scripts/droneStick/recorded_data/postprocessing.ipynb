```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
from numpy.linalg import norm
from math import *
```

```python
def euler_from_quaternion(q):
    """
    Intrinsic Tait-Bryan rotation of xyz-order.
    """
    q = np.array(q)
    q = q / np.linalg.norm(q)
    qx, qy, qz, qw = q
    roll = atan2(2.0*(qy*qz + qw*qx), qw*qw - qx*qx - qy*qy + qz*qz)
    pitch = asin(-2.0*(qx*qz - qw*qy))
    yaw = atan2(2.0*(qx*qy + qw*qz), qw*qw + qx*qx - qy*qy - qz*qz)
    return roll, pitch, yaw
```

```python
class Experiment:
    def __init__(self,):
        self.joystick = None
        self.drone = None
        self.drone_sp = None
        self.subject_name = ''
        self.obstacle = None
        self.path = ''
        
    def load_data(self):
        self.joystick = pd.read_csv( self.path + '_slash_vicon_slash_cf4_slash_cf4.csv')
        self.drone = pd.read_csv(self.path + '_slash_vicon_slash_cf2_slash_cf2.csv')
        self.drone_sp = pd.read_csv(self.path + '_slash_cf2_sp.csv')
        self.obstacle = pd.read_csv(self.path + '_slash_vicon_slash_obstacle25_slash_obstacle25.csv')
        
    def plot_traj(self, t_ind_limit=None):
        if t_ind_limit == None: t_ind_limit = len(self.drone.x)
        plt.plot(self.drone.x[:t_ind_limit], self.drone.y[:t_ind_limit], linewidth=5)
        plt.grid()
```

```python
def get_angles(joystick):
    roll_array = []; pitch_array = []; yaw_array = []
    for i in range(len(joystick['x.1'])):
        q = [joystick['x.1'][i], joystick['y.1'][i], joystick['z.1'][i], joystick['w'][i]]
        roll, pitch, yaw = euler_from_quaternion(q)
        roll_array.append(roll); pitch_array.append(pitch); yaw_array.append(yaw)

    roll_array = np.array(roll_array)
    pitch_array = np.array(pitch_array)
    t = np.array( joystick.secs ) + np.array(joystick.nsecs / 1000000000)
    t -= t[0]
    
    plt.plot(roll_array)

    return roll_array, pitch_array, yaw_array, t
```

```python
r_short, p_short, y_short, t_short = get_angles(Batyr_short.joystick)
```

```python
plt.figure(figsize=(8,6))

plt.plot(t_long, roll_long, label='long rope')
plt.plot(t_short, r_short, label='short rope')
plt.title("Roll")
plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(8,6))

plt.plot(t_long, pitch_long, label='long rope')
plt.plot(t_short, p_short, label='short rope')
plt.title("Pitch")
plt.grid()
plt.legend()
```
