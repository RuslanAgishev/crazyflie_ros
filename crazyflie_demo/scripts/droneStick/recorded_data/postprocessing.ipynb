```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
from numpy.linalg import norm
from math import *
```

```python
def euler_from_quaternion(q):
    """
    Intrinsic Tait-Bryan rotation of xyz-order.
    """
    q = np.array(q)
    q = q / np.linalg.norm(q)
    qx, qy, qz, qw = q
    roll = atan2(2.0*(qy*qz + qw*qx), qw*qw - qx*qx - qy*qy + qz*qz)
    pitch = asin(-2.0*(qx*qz - qw*qy))
    yaw = atan2(2.0*(qx*qy + qw*qz), qw*qw + qx*qx - qy*qy - qz*qz)
    return roll, pitch, yaw
```

# Alexander

```python
name = 'Alexander'
path = root_path + name + '/data_2019-04-03-13-02-34/'

joystick = pd.read_csv( path + '_slash_vicon_slash_cf4_slash_cf4.csv')
drone = pd.read_csv(path + '_slash_vicon_slash_cf2_slash_cf2.csv')
drone_sp = pd.read_csv(path + '_slash_cf2_sp.csv')
obstacle = pd.read_csv(path + '_slash_vicon_slash_obstacle25_slash_obstacle25.csv')
```

```python
joystick.head()
```

```python
joystick.head()
```

```python
roll_array = []; pitch_array = []; yaw_array = []
for i in range(len(joystick['x.1'])):
    q = [joystick['x.1'][i], joystick['y.1'][i], joystick['z.1'][i], joystick['w'][i]]
    roll, pitch, yaw = euler_from_quaternion(q)
    roll_array.append(roll); pitch_array.append(pitch); yaw_array.append(yaw)
    
roll_array = np.array(roll_array)
pitch_array = np.array(pitch_array)
```

```python
len(roll_array)
```

```python
t = np.array( joystick.secs ) + np.array(joystick.nsecs / 1000000000)
t -= t[0]
t_ind_limit = 4100
plt.figure(figsize=(8,6))
plt.plot(t[:t_ind_limit], roll_array[:t_ind_limit], label='roll')
plt.plot(t[:t_ind_limit], pitch_array[:t_ind_limit], label='pitch')
# plt.plot(t[:4100], yaw_array[:4100])
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=(10,10))
plt.plot(joystick.x[:t_ind_limit], joystick.y[:t_ind_limit], linewidth=2)
plt.plot(drone.x[:t_ind_limit], drone.y[:t_ind_limit], linewidth=5)
plt.plot(obstacle.x[0], obstacle.y[0], 'ro', markersize=30, color='k')
plt.grid()
```

```python
vel_coef = 7.0

plt.figure(figsize=(8,6))
plt.plot(t[:t_ind_limit], vel_coef * roll_array[:t_ind_limit], label='7 x roll')
plt.plot(t[:t_ind_limit], drone.y[:t_ind_limit], label='drone X, [m]')
plt.legend()
plt.grid()
```

```python
vel_coef = 7.0

plt.figure(figsize=(8,6))
plt.plot(t[:t_ind_limit], vel_coef * pitch_array[:t_ind_limit], label='7 x pitch')
plt.plot(t[:t_ind_limit], drone.x[:t_ind_limit], label='drone Y, [m]')
plt.legend()
plt.grid()
```
